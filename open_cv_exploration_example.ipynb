## Using OpenCv as part of preprocessing step

Using openCv for preprocessing images before feeding them into a CNN can be very useful and, in many cases, can improve the accuracy of the model. 
<br>

Below are several elementary examples. We recommend experimenting with your own!

## Adaptive thresholding

```julia
labels = list(set(train_data['label']))

plt.figure(figsize=(10,10))
for i, label in enumerate(labels):
    df = train_data[train_data['label'] == label].reset_index()
    img = cv2.imread(Config.images_path + '/' +  df['image'][0])
    plt.subplot(5, 5, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    th3 = cv2.adaptiveThreshold(img,128,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2)
    plt.imshow(th3,'gray')
    plt.title(label)
plt.show()
```

## CV2 canny edge detection

```julia
labels = list(set(train_data['label']))

plt.figure(figsize=(10,10))
for i, label in enumerate(labels):
    df = train_data[train_data['label'] == label].reset_index()
    img = cv2.imread(Config.images_path + '/' +  df['image'][0])
    plt.subplot(5, 5, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    #img = cv2.imread(train_images[i])
    #img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    th3 = cv2.Canny(img,100,500)
    plt.imshow(th3,'gray')
    plt.title(label)
plt.show()
```

```julia
labels = list(set(train_data['label']))

plt.figure(figsize=(10,10))
for i, label in enumerate(labels):
    df = train_data[train_data['label'] == label].reset_index()
    img = cv2.imread(Config.images_path + '/' +  df['image'][0])
    plt.subplot(5, 5, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
    thresh1 = thresh1 / 255
    plt.imshow(thresh1)
    plt.title(label)
plt.show()
```
