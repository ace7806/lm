```python
import pandas as pd
import torch 
import torch.nn as nn
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from torchvision import transforms
```

# What is Image Pre-Processing/Augmentation?

<img src="images/workflow_of_machine_learning.jpeg" width="450"/>

<b>Data preprocessing</b> is the step within the data science lifecycle of cleaning, transforming, and integrating data, in this case images, to make it ready for analysis and modeling. It plays a crucial role in improving performance and robustness of machine learning models moving forward. For this mini-session we'll be focusing on image augmentation methods.

<b>Image augmentation</b> specifically focuses on generating new training examples by applying random transformations to the input images. Our objective here is to simulate real-world variability into the image set and enhance the model's ability to generalize. 

## Types of Image Augmentations
Let's go through some of the more common techniques:
- ### Geometric Transformations
    - <b>Translation</b>: Shifting the image horizontally and/or vertically
    - <b>Scaling/Zooming</b>: Enlarging or shrinking the image
    - <b>Elastic Transform</b>
    - <b>Padding</b>

```python
#Load up the original image
img = Image.open('/mnt/cnn_session/road_img.png')
img
```

```python
geom_transform = [
    ("Original", transforms.Compose([])), #Original road image
    ("Translation", transforms.Compose([transforms.RandomAffine(0, translate=(0.4,0.4))])), #Shift the image by up to 40% of image size
    ("Scaling/Zooming", transforms.Compose([transforms.RandomAffine(0, scale=(1.5, 5.3))])), #Scaling between 150% and 530%
    ("Elastic Transform", transforms.Compose([transforms.ElasticTransform(alpha=250.0)]))
]

#Don't forget about padding!
geom_transform.append(("Padding", transforms.Compose([transforms.Pad(50)]))) #Adding 50 pixels of padding)
```

*Let's plot it to see how the image has been affected:*

```python
fig, axes = plt.subplots(1,len(geom_transform), figsize=(40,5))

for i, (transform_type, transformation) in enumerate(geom_transform):
    transformed_image = transformation(img)
    
    axes[i].imshow(transformed_image)
    axes[i].set_title(transform_type, fontsize=25)
    axes[i].axis('off')

plt.show()
```

- ### Color Adjustments
    - <b>Gaussian Blur</b>
    - <b>Random Invert</b>
    - <b>Greyscale</b>
    - <b>Color Jitter</b>

```python
color_transform = [
    ("Original", transforms.Compose([])), #Original road image
    ("Gaussian Blur", transforms.Compose([transforms.GaussianBlur(kernel_size=(5, 13), sigma=(0.1, 5))])),
    ("Random Invert", transforms.Compose([transforms.RandomInvert()])),
    ("Greyscale", transforms.Compose([transforms.Grayscale(num_output_channels=3)])),
    ("Color Jitter", transforms.Compose([transforms.ColorJitter(brightness=0.2,contrast=0.2, saturation=0.2,hue=0.2)]))
]
```

```python
fig, axes = plt.subplots(1,len(geom_transform), figsize=(40,5))

for i, (transform_type, transformation) in enumerate(color_transform):
    transformed_image = transformation(img)
    
    axes[i].imshow(transformed_image)
    axes[i].set_title(transform_type, fontsize=25)
    axes[i].axis('off')

plt.show()
```
