```python
import pandas as pd
import torch 
import matplotlib.pyplot as plt
import cv2
import numpy as np
from torchvision.transforms import transforms
```

## Data Augmentation

### Intro
>Data augmentation allows us to effectively deal with cases where the training data is limited. Certain techniques will allow us to modify certain images and use them as "new" training data to boost the performance of our model. The advantage of using data augmentation techniques is that the model will view augmented data as new data. For example, as far the CNN is concerned a horizontally flipped image is a different input than the original image. Capitalizing on this trick will allow you to take your model performance to the next level. 

### Techniques 
> Just a few of the techniques include shifting your image horizontally or vertically, zooming in or out, altering brightness, etc. So whether you want to combat data imbalance or simply provide a larger set of training data to your model, this may be a great topic to look into! [TRANSFORMING AND AUGMENTING IMAGES](https://pytorch.org/vision/stable/transforms.html) is the official PyTorch documentation, and it can guide you through this process. 

```python
train_data = pd.read_csv('/domino/datasets/UPRM_Hackathon_2023/train_data_uprm.csv')
```

```python
class Config: 
    images_path = '/domino/datasets/UPRM_Hackathon_2023/'
```

```python
sample_classes = [7, 22, 23]

images = []
for i, label in enumerate(sample_classes):
    df = train_data[train_data['label'] == label].reset_index()
    img = cv2.imread(Config.images_path + '/' +  df['image'][1])
    images.append(img)
```

```python
torch.manual_seed(1)
fig = plt.figure(figsize=(14,12))
for j, img in enumerate(images):
    ax = fig.add_subplot(3,4,j*4+1)
    ax.imshow(img[:,:,::-1])
    if j == 0:
        ax.set_title('Original', size=15)
        
    ax = fig.add_subplot(3,4,j*4+2)
    img_transform = transforms.Compose([
        transforms.RandomCrop([100,100])
    ])
    img = np.transpose(img, (2,0,1))
    img = torch.from_numpy(img)
    img_cropped = img_transform(img)
    img_cropped = np.transpose(img_cropped, (1,2,0))
    ax.imshow(img_cropped[:,:,[2,1,0]])
    
    if j == 0:
        ax.set_title('Random Crop', size=15)
        
    ax = fig.add_subplot(3,4,j*4+3)
    img_transform = transforms.Compose([
        transforms.RandomHorizontalFlip(0.8), # probability of flip is normally set to 0.5
        transforms.RandomVerticalFlip(0.8)
    ])
    img_flipped = img_transform(img)
    img_flipped= np.transpose(img_flipped, (1,2,0))
    ax.imshow(img_flipped[:,:,[2,1,0]])
    
    if j == 0:
        ax.set_title('Random flip', size=15)
    
    
plt.show()
```
